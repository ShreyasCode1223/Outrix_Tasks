```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset. Make sure you've uploaded it to Colab's session storage.
file_path = 'StudentsPerformance.csv'
df = pd.read_csv(file_path)

print("First 5 rows of the dataset:")
print(df.head())
print("\nDataFrame Info:")
df.info()
```

```python
# Check for any missing values
print("\nMissing values check:")
print(df.isnull().sum())

# Create a new feature: 'average score'
df['average score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
df['average score'] = df['average score'].round(2)

print("\nFirst 5 rows with the new 'average score' column:")
print(df.head())
```

```python
# Plot 1: Distribution of Average Scores
plt.figure(figsize=(10, 6))
sns.histplot(df['average score'], bins=15, kde=True)
plt.title('Distribution of Student Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Average Scores by Parental Level of Education
plt.figure(figsize=(12, 7))
sns.barplot(x='parental level of education', y='average score', data=df, errorbar=None)
plt.title('Average Score by Parental Education Level')
plt.xticks(rotation=45, ha='right')
plt.show()

# Plot 3: Average Scores by Lunch Type
plt.figure(figsize=(8, 6))
sns.boxplot(x='lunch', y='average score', data=df)
plt.title('Average Score by Lunch Type')
plt.show()

# Plot 4: Average Scores by Test Preparation Course
plt.figure(figsize=(8, 6))
sns.violinplot(x='test preparation course', y='average score', data=df)
plt.title('Average Score by Test Preparation Course Completion')
plt.show()
```
